# DATA CLEANING

```python
df_recipe = pd.read_csv('RecipeRecommendation_nocal.csv')
```

```python
#Menambahkan kolom Calories
df_recipe['Calories'] = 0
```

```python
df_recipe
```

```python
#Memasukkan beberapa value Calorie Makanan
calories_dict = {
    'Ayam Woku Manado': 350,
    'Ayam goreng tulang lunak': 400,
    'Ayam cabai kawin': 380,
    'Ayam Geprek': 450,
    'Minyak Ayam': 120
}

df_recipe['Calories'] = df_recipe['Title'].map(calories_dict)
```

```python
total_rows = len(df_recipe)
mid_point = total_rows // 2

#Mengisi setengah data awal dengan random value (150 - 499)
df_recipe.loc[:mid_point, 'Calories'] = df_recipe.loc[:mid_point, 'Calories'].apply(
    lambda x: np.random.randint(150, 500) if pd.isna(x) else x
)

#Mengisi setengah data awal dengan random value (500 - 850)
df_recipe.loc[mid_point:, 'Calories'] = df_recipe.loc[mid_point:, 'Calories'].apply(
    lambda x: np.random.randint(500, 851) if pd.isna(x) else x
)


# Mengubah tipe data kolom 'Calories' menjadi integer
df_recipe['Calories'] = df_recipe['Calories'].astype(int)
```
